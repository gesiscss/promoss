```python
%run 'hmdp.ipynb'

#directory of meta.txt and corpus.txt
directory = "/home/c/work/topicmodels/food_test/";
#the first value in meta.txt are Geographical coordinates
#and we want to detect 1000 clusters
meta_params = "G(1000)";

#construct HMDP class
hmdp = HMDP(directory, meta_params);
#set max. topics which can be detected
hmdp.T = 25;
#we only keep words which appear at least the given number of times
hmdp.MIN_DICT_WORDS = 2;
#run model for a given number of iterations
hmdp.RUNS = 100;

#run model
hmdp.run();

hmdp.map_from_JSON();
```
